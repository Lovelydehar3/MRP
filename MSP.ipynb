```python
import ast
from collections import Counter

# ‚úÖ Safely parse 'cast' field
def safe_eval_cast(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    elif isinstance(x, list):
        return x
    else:
        return []

movies['cast'] = movies['cast'].apply(safe_eval_cast)

# üé≠ Extract top 3 actors
def extract_actors(cast_list):
    return [member['name'] for member in cast_list[:3]]

movies['top_actors'] = movies['cast'].apply(extract_actors)

# Flatten all actors and count
all_actors = sum(movies['top_actors'], [])
actor_counts = Counter(all_actors).most_common(10)
actor_df = pd.DataFrame(actor_counts, columns=['Actor', 'Count'])

# üìä Barplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.barplot(data=actor_df, x='Actor', y='Count', hue='Actor', palette='cubehelix', legend=False)
plt.title("Top 10 Most Frequent Actors")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (replace with your correct path)
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")

# Merge on 'title'
df = pd.merge(ratings, movies, on='title')

# ------------ Feature 1: Genre Count ------------ #
df['genres'] = df['genres'].fillna('')
df['genres'] = df['genres'].apply(lambda x: x.replace('[','').replace(']','').replace("'",'').replace(" ",''))
genres_list = df['genres'].str.split(',')
all_genres = [genre for sublist in genres_list for genre in sublist if genre != '']
genre_df = pd.Series(all_genres).value_counts().reset_index()
genre_df.columns = ['Genre', 'Count']

plt.figure(figsize=(8,5))
sns.barplot(data=genre_df.head(10), x='Genre', y='Count')
plt.title('Top 10 Genres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------ Feature 2: Original Language ------------ #
lang_df = df['original_language'].value_counts().head(10)

plt.figure(figsize=(8,5))
lang_df.plot(kind='bar', color='skyblue')
plt.title('Top 10 Languages')
plt.xlabel('Language')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()

# ------------ Feature 3: Budget vs Revenue ------------ #
df_budget = df[(df['budget'] > 0) & (df['revenue'] > 0)]

plt.figure(figsize=(8,5))
plt.scatter(df_budget['budget'], df_budget['revenue'], alpha=0.5)
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()

# ------------ Feature 4: Runtime Distribution ------------ #
plt.figure(figsize=(8,5))
plt.hist(df['runtime'].dropna(), bins=30, color='orange', edgecolor='black')
plt.title('Runtime Distribution')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```
